```python
import pandas as pd
```

```python
df_1 = pd.read_csv("./data/leetcode_questions.csv")
df_2 = pd.read_csv("./data/leetcode_solutions.csv")
```

```python
df_merged = pd.merge(df_1, df_2, left_on='Question ID', right_on='id', how='inner')
```

```python
df_cleaned = df_merged.drop(columns=["Question Text", "Question ID", "Question Title", "Success Rate", "total submission", "total accepted", "Likes", "Dislikes", "Similar Questions Text", "Similar Questions ID"])
df_cleaned.head()
```

```python
df_exploded = df_cleaned.assign(Topic_Tagged=df_cleaned['Topic Tagged text'].str.split(',')).explode('Topic_Tagged')
unique_topics = sorted(df_exploded['Topic_Tagged'].str.strip().dropna().unique())
```

```python
df_filtered = df_cleaned[
    df_cleaned['Topic Tagged text'].str.contains(r'\bTwo Pointers\b', case=False, na=False) & 
    (df_cleaned['Difficulty Level'].str.lower() == 'easy')
]
pd.set_option('display.max_colwidth', None)
print(df_filtered[:1])
```
